```python
#Importamos solo las librer铆as necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#  Simulamos datos (puedes cambiarlos si quieres usar reales)
np.random.seed(0)
cantidad_palabras = np.random.randint(300, 2500, 100)
compartidos = cantidad_palabras * 20 + np.random.randint(0, 50000, 100)

#  Preparamos los datos para el modelo
X = cantidad_palabras.reshape(-1, 1)  # Variable independiente
y = compartidos  # Variable dependiente

#  Creamos y entrenamos el modelo de regresi贸n
modelo = LinearRegression()
modelo.fit(X, y)

#  Visualizamos los datos y la recta de regresi贸n
plt.scatter(X, y, color='blue', label='Datos Reales')
plt.plot(X, modelo.predict(X), color='red', linewidth=2, label='Recta de Regresi贸n')
plt.title("Regresi贸n Lineal")
plt.xlabel("Cantidad de Palabras")
plt.ylabel("Compartidos")
plt.legend()
plt.grid(True)
plt.show()

# М Predicci贸n con entrada por teclado
try:
    cantidad_input = float(input(" Ingresa la cantidad de palabras para predecir los compartidos: "))
    prediccion = modelo.predict([[cantidad_input]])
    print(f" La predicci贸n de compartidos para {cantidad_input} palabras es: {prediccion[0]:.2f}")
except ValueError:
    print("锔 Por favor, ingresa un n煤mero v谩lido.")
```
