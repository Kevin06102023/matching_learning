```python
#Importamos solo las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Datos simulados: cantidad de palabras y veces compartido
palabras = np.array([200, 300, 400, 500, 600, 700, 800]).reshape(-1, 1)
compartidos = np.array([20, 25, 30, 35, 42, 48, 55])

# Creamos y entrenamos el modelo
modelo = LinearRegression()
modelo.fit(palabras, compartidos)

# Pedimos al usuario que ingrese una cantidad de palabras
try:
    entrada = float(input("Ingresa la cantidad de palabras para predecir los compartidos: "))
    prediccion = modelo.predict(np.array([[entrada]]))
    print(f"\nPredicción: Un artículo de {int(entrada)} palabras será compartido aproximadamente {int(prediccion[0])} veces.")
    
    # Visualización
    plt.scatter(palabras, compartidos, color='blue', label='Datos reales')
    plt.plot(palabras, modelo.predict(palabras), color='red', label='Línea de regresión')
    plt.scatter(entrada, prediccion, color='green', label='Predicción')
    plt.xlabel("Cantidad de palabras")
    plt.ylabel("Veces compartido")
    plt.title("Predicción de compartidos según cantidad de palabras")
    plt.legend()
    plt.grid(True)
    plt.show()

except ValueError:
    print("Por favor, ingresa un número válido.")
```
